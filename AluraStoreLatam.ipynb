### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Sumamos la columna 'Precio' para cada DataFrame
facturacion_tienda1 = tienda['Precio'].sum()
facturacion_tienda2 = tienda2['Precio'].sum()
facturacion_tienda3 = tienda3['Precio'].sum()
facturacion_tienda4 = tienda4['Precio'].sum()

print(f"Facturación Tienda 1: ${facturacion_tienda1:,.2f}")
print(f"Facturación Tienda 2: ${facturacion_tienda2:,.2f}")
print(f"Facturación Tienda 3: ${facturacion_tienda3:,.2f}")
print(f"Facturación Tienda 4: ${facturacion_tienda4:,.2f}")
```

# 2. Ventas por categoría

```python
# Agrupamos por categoría y contamos las ocurrencias
ventas_cat1 = tienda['Categoría del Producto'].value_counts()
ventas_cat2 = tienda2['Categoría del Producto'].value_counts()
ventas_cat3 = tienda3['Categoría del Producto'].value_counts()
ventas_cat4 = tienda4['Categoría del Producto'].value_counts()

print("Top Categoría Tienda 1:", ventas_cat1.idxmax())
print("Top Categoría Tienda 2:", ventas_cat2.idxmax())
print("Top Categoría Tienda 3:", ventas_cat3.idxmax())
print("Top Categoría Tienda 4:", ventas_cat4.idxmax())
```

# 3. Calificación promedio de la tienda


```python
# Promedio de la columna 'Calificación'
media_tienda1 = tienda['Calificación'].mean()
media_tienda2 = tienda2['Calificación'].mean()
media_tienda3 = tienda3['Calificación'].mean()
media_tienda4 = tienda4['Calificación'].mean()

print(f"Calificación promedio Tienda 1: {media_tienda1:.2f}")
print(f"Calificación promedio Tienda 2: {media_tienda2:.2f}")
print(f"Calificación promedio Tienda 3: {media_tienda3:.2f}")
print(f"Calificación promedio Tienda 4: {media_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
# Usamos una función simple para mostrar los extremos de cada tienda
def productos_extremos(df, nombre):
    conteo = df['Producto'].value_counts()
    print(f"--- {nombre} ---")
    print(f"Más vendido: {conteo.idxmax()} ({conteo.max()})")
    print(f"Menos vendido: {conteo.idxmin()} ({conteo.min()})")

productos_extremos(tienda, "Tienda 1")
productos_extremos(tienda2, "Tienda 2")
productos_extremos(tienda3, "Tienda 3")
productos_extremos(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
# Promedio de la columna 'Costo de envío'
envio_tienda1 = tienda['Costo de envío'].mean()
envio_tienda2 = tienda2['Costo de envío'].mean()
envio_tienda3 = tienda3['Costo de envío'].mean()
envio_tienda4 = tienda4['Costo de envío'].mean()

print(f"Envío promedio Tienda 1: ${envio_tienda1:,.2f}")
print(f"Envío promedio Tienda 2: ${envio_tienda2:,.2f}")
print(f"Envío promedio Tienda 3: ${envio_tienda3:,.2f}")
print(f"Envío promedio Tienda 4: ${envio_tienda4:,.2f}")
```

Visualizacion

```python
import matplotlib.pyplot as plt

nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
facturaciones = [facturacion_tienda1, facturacion_tienda2, facturacion_tienda3, facturacion_tienda4]
calificaciones = [media_tienda1, media_tienda2, media_tienda3, media_tienda4]
costos_envio = [envio_tienda1, envio_tienda2, envio_tienda3, envio_tienda4]

fig, ax = plt.subplots(1, 3, figsize=(18, 5))

# Gráfico 1: Barras (Facturación)
ax[0].bar(nombres, facturaciones, color='teal')
ax[0].set_title('Facturación Total ($)')

# Gráfico 2: Línea (Calificaciones)
ax[1].plot(nombres, calificaciones, marker='o', color='orange')
ax[1].set_title('Calificación Promedio (1-5)')
ax[1].set_ylim(0, 5)

# Gráfico 3: Pastel (Distribución de Costos de Envío)
ax[2].pie(costos_envio, labels=nombres, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
ax[2].set_title('Proporción de Costos de Envío')

plt.tight_layout()
plt.show()
```
